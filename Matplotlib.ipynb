```python
age = list(range(18,25))
bcount = [1,2,4,7,8,10,15]
plt.figure(figsize=(6,6),dpi=150)
plt.bar(age,bcount,color='red',width=0.2)
```

```python
year = list(range(1990,1999))

unemp_rate = [13,15,19,9,8,7,2,3,1]
plt.title("Unemployment across the years")
plt.xlabel("years")
plt.ylabel("unemployment rate")
plt.bar(year,unemp_rate,color="red",width=0.2)
```

```python
Year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
Unemployment_Rate = [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]

plt.figure(figsize=(8,6),dpi=100)
plt.title('Unemployment Rate Vs Year', fontsize=15)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Unemployment Rate', fontsize=13)
plt.grid(True)
plt.plot(Year, Unemployment_Rate, color='red', marker='X')
plt.show()
```

```python
country = ['India','Japan',"USA","Russia","Italy","Germany"]
avgage = [25,49,39,41,46.4,44.7]
new_col = ['blue','orange','green','red','purple','maroon']
plt.grid(True)
plt.title("Average Age for Countries",fontsize=15)
plt.xlabel("Countries")
plt.ylabel("Average Age",fontsize=10)
plt.plot(country,avgage,marker='x',markersize=12,linewidth=2,color='teal')
plt.bar(country,avgage,color=new_col)
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example one")
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example two", color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Wow! We Got Our First Bar Graph')
plt.show()
```
