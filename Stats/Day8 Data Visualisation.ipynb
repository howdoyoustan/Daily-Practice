```python
df = pd.read_excel(r'E:\dbda khargar\Stats and R\data_visual.xlsx',sheet_name='BarChart')
```

```python
df
```

```python
cat = df.Category
cnt = df.Counts
```

```python
#In both histogram and barchart their y-axis is frequency
#incase of x-axis for barchart they are categorical data, they do not touch each other, bars are of equal thickness
#incase of x-axis for histogram they are continuous (quantitative) data, they might touch each other and their thickness maybe different 
plt.bar(cat,cnt)
plt.xlabel('Category')
plt.ylabel('Counts')
```

```python
plt.barh(cat,cnt)
plt.xlabel('Category')
plt.ylabel('Counts')
```

```python
plt.pie(cnt,labels=cat)
```
